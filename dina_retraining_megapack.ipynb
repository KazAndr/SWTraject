Файл для тренеровки нейронной сети на размноженном наборе данных

```python
import os 
import glob
import datetime

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.externals import joblib
```

```python
gp_table = pd.read_table('typized_dataset_for_dina_megaset_2020-03-11.csv', sep='\t')
final_test = pd.read_table('typized_dataset_for_dina_final_test_2020-03-11.csv', sep='\t')
```

```python
gp_table
```

```python
final_test
```

```python
X = gp_table.drop(['type'], axis=1)
y = gp_table['type']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
clf = RandomForestClassifier(n_estimators=1000)
clf = clf.fit(X_train, y_train)
predictions = clf.predict(X_test)
print(accuracy_score(predictions, y_test))
```

```python
# проверка на внешнем наборе не тронутом размноживанием
X_valid = final_test.drop(['type'], axis=1)
y_valid = final_test['type']
predictions = clf.predict(X_valid)
print(accuracy_score(predictions, y_valid))
```

```python
# Сохранение классификатора
NOW = datetime.datetime.now().strftime("%Y-%m-%d")
filename = f'dina_model_RFC(1000)_megaset_{NOW}.sav'
joblib.dump(clf, filename)
```
