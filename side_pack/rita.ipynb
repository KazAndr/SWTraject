```python
head, main_pulse, data_pulses, back = read_profiles('171118_1133+16_00_profiles.txt')
```

```python
non_cor_data = []
for pulse, backg in zip(data_pulses, back):
    non_cor_data.append(pulse + backg)
```

```python
obser = np.hstack(non_cor_data)
```

```python
plt.close()
plt.plot(obser)
plt.show()
```

```python
diag = []
for x in tqdm_notebook(np.linspace(-1.2, 1.2, len(obser))):
    if x == 0:
        diag.append(1)
    else:
        diag.append((np.sin(x)/x)**2)
```

```python
cor_d = []
for data_point, coeff in zip(obser, diag):
    cor_d.append(data_point/coeff)
cor_d = np.asarray(cor_d)
```

```python
plt.close()
plt.plot(cor_d)
plt.show()
```
