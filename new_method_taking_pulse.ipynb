```python
files_0531 = sorted(
    glob.glob('./final_dataset/' + os.sep + 'obs_data' + os.sep + '*'),
    key=lambda x: datetime.datetime.strptime(os.path.basename(x), '%d.%m.%Y_obs_0531+21.csv'))
```

```python
flat_obs = np.genfromtxt(files_0531[3], skip_header=7)
```

```python
plt.close()
plt.plot(flat_obs)
#plt.xlim(51500, 51850)
plt.show()
```

```python
test_flat_obser = deepcopy(flat_obs)
med_flux = np.median(test_flat_obser)
```

```python
loaded_model = joblib.load('dina_model_RFC_pulse_noise_95.sav')
```

```python
skip_index = []
i = 0
while np.max(test_flat_obser) >= 1725 and np.argmax(np.max(test_flat_obser)) not in skip_index:
    x_max = np.argmax(test_flat_obser)
    pulse = test_flat_obser[x_max - 10: x_max + 90] - med_flux
    n_pulse = pulse/max(pulse)
    NN_decition = loaded_model.predict([n_pulse])
    if NN_decition[0] == 1:
        medias = np.full(len(pulse), med_flux)
        test_flat_obser[x_max - 10: x_max + 90] = medias
        plt.close()
        plt.plot(pulse)
        plt.savefig(f'./final_dataset/gp_plot/{i}.png', format='png')
        skip_index.append(x_max)
        i += 1
    else:
        skip_index.append(x_max)
```
