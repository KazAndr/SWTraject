```python
import os
import sys
import glob
import platform
import datetime

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from tqdm import tqdm_notebook
from astropy import units as u
from astropy.time import Time
```

```python
gp_table = pd.read_table('crab_gp_kaz_10_2010-2018_calib_sigma.csv', sep='\t')
```

```python
gp_table.head()
```

```python
gp_table['MJD_start'] = None
gp_table['MJD_stop'] = None
```

```python
for index, row in gp_table.iterrows():
    day, month, year = row['Date'].split('.')
    time_utc_start = (Time('{}-{}-{}T'.format(year, month, day) + row['Time start'], precision=7)
            - 4*u.hour)
    time_utc_stop = (time_utc_start
                     + (int(row['Numpulse, a.u.']) - 1)
                     *int(row['Numpointwin, point'])
                     *float(row['Tay, ms'])*u.millisecond)
    gp_table.loc[index,'MJD_start'] = str(time_utc_start.mjd)
    gp_table.loc[index,'MJD_stop'] = str(time_utc_stop.mjd)
```

## Заполнение таблицы сеансов исходя из директории obs_data

```python
files_0531 = sorted(
    glob.glob('.' + os.sep + 'obs_data' + os.sep + '*'),
    key=lambda x: datetime.datetime.strptime(os.path.basename(x), '%d.%m.%Y_obs_0531+21.csv'))

def read_head(filename, numpar):
    header = {}
    with open(filename, 'r') as file:
        for i in range(numpar):
            a, *b = file.readline().split()
            try:
                header[a] = b[0] + '.' + b[1]
            except IndexError:
                header[a] = b[0].replace(',', '.')
    return header
```

```python
crab_sessions = pd.DataFrame(columns=[
    'Date',
    'MJD_start',
    'MJD_stop',
    'Session'])
```

```python
for idx, name in enumerate(files_0531):
    head = read_head(name, 7)
    day, month, year = head['date'].split('.')
    time_utc_start = (Time('{}-{}-{}T'.format(year, month, day) + head['time'], precision=7)
            - 4*u.hour)
    time_utc_stop = (time_utc_start
                     + (int(head['numpuls']) - 1)
                     *int(head['numpointwin'])
                     *float(head['tay'])*u.millisecond)
    crab_sessions.loc[idx] = [
        head['date'],
        str(time_utc_start.mjd),
        str(time_utc_stop.mjd),
        1,
    ]
```

```python
crab_sessions
```

```python
crab_sessions.to_csv(
        'crab' + '_obs_kaz_MJD.csv',  sep='\t', header=True, index=False)
```
