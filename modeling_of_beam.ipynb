```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def beam(x):
    return (np.sin(x)/x)*(np.sin(x)/x)

def gauss(x,a,x0,sigma):
    return a*np.exp(-(x-x0)**2/(2*sigma**2))

def meander(x):
    if -0.05 <= x <= 0.05:
        return 1
    else:
        return 0
```

```python
x = np.linspace(-np.pi, np.pi, 100)
```

```python
plt.clf()
plt.plot(x, beam(x))
plt.show()
```

Ширина диаграммы направленности по уровню 0.5 состовляет 3 еденицы. 
Ширина туманности, соответственно в 30 раз меньше, т.е. 0.1

```python
plt.clf()
plt.plot(x, gauss(x,1,0,0.05))
plt.show()
```

```python
plt.clf()
plt.plot(x, [meander(i) for i in x])
plt.show()
```

Свертка с диаграммой:

```python
convole_gauss = np.convolve(beam(x), gauss(x,1,0,0.05), 'same')/max(np.convolve(beam(x), gauss(x,1,0,0.05)))
convole_meandr = np.convolve(beam(x), [meander(i) for i in x], 'same')/max(np.convolve(beam(x),[meander(i) for i in x]))
```

```python
plt.clf()
plt.plot(x, beam(x), '--', label='(sin(x)/x)^2')
plt.plot(x, convole_gauss, label='beam + gauss')
plt.plot(x, convole_meandr, label='beam + meandr')
plt.legend()
plt.savefig(f'./some_images/modeling_of_beam.png', format='png', dpi=300)
```
