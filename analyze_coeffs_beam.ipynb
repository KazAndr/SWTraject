```python
import os
import sys
import glob
import shutil
import datetime

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
coeff_table = pd.read_table('beam_coeff_obs_kaz.csv', sep='\t')
```

```python
coeff_table
```

```python
coeff_table['Time_series'] = pd.to_datetime(coeff_table['Date'], format='%d.%m.%Y', errors='ignore')
coeff_table = coeff_table.set_index('Time_series')
```

```python
plt.close()
plt.plot(coeff_table['A'])
plt.ylabel('Coeff A, ADC units')
plt.grid()
plt.plot()
```

```python
plt.close()
plt.plot(coeff_table['shift'])
plt.ylabel('Shift, points')
plt.grid()
plt.plot()
```

```python
plt.close()
plt.plot(coeff_table['y0'])
plt.ylabel('Y0, ADC units')
plt.grid()
plt.plot()
```

```python
plt.close()
plt.plot(coeff_table['Jy/ADC'])
plt.ylabel('Jy/ADC')
plt.grid()
plt.plot()
```
