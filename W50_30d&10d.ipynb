```python
import os
import sys
import platform

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
gp_table = pd.read_table('crab_gp_kaz_pulses.csv', sep='\t')
```

```python
gp_table.head(30)
```

```python
resul_tab = pd.DataFrame(columns=[
    'MJD_mean',
    'W50_mean',
    'STD'])
```

```python
srez_mjd = np.asarray(gp_table['MJD'])
srez_w50 = np.asarray(gp_table['W50, point'])
```

```python
from math import ceil

def chunk(arr, size):
    return list(
        map(lambda x: arr[x * size:x * size + size],
            list(range(0, ceil(len(arr) / size)))))
```

```python
srez_mjd_30 = chunk(srez_mjd, 10)
srez_w50_30 = chunk(srez_w50, 10)
```

```python
srez_mjd_30[-1]
```

```python
srez_w50_30[-1]
```

```python
idx = 0
for mjd, w50 in zip(srez_mjd_30, srez_w50_30):
    resul_tab.loc[idx] = [
            np.mean(mjd),
            np.mean(w50),
            np.std(w50)]
    idx +=1
```

```python
resul_tab
```

```python
resul_tab.to_csv('W50_bin_10.csv', sep='\t', header=True, index=False)
```
