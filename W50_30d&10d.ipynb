```python
import os
import sys
import platform

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

from astropy.time import Time
```

```python
gp_table = pd.read_table('crab_gp_kaz_pulses_sigma.csv', sep='\t')
```

```python
gp_table.head(30)
```

```python
resul_tab = pd.DataFrame(columns=[
    'MJD_mean',
    'W50_mean',
    'STD'])
```

```python
srez_mjd = np.asarray(gp_table['MJD'])
srez_w50 = np.asarray(gp_table['W50, point'])
```

```python
from math import ceil

def chunk(arr, size):
    return list(
        map(lambda x: arr[x * size:x * size + size],
            list(range(0, ceil(len(arr) / size)))))
```

```python
srez_mjd_30 = chunk(srez_mjd, 10)
srez_w50_30 = chunk(srez_w50, 10)
```

```python
srez_mjd_30[-1]
```

```python
srez_w50_30[-1]
```

```python
idx = 0
for mjd, w50 in zip(srez_mjd_30, srez_w50_30):
    resul_tab.loc[idx] = [
            np.mean(mjd),
            np.mean(w50),
            np.std(w50)]
    idx +=1
```

```python
resul_tab
```

```python
resul_tab.to_csv('W50_bin_10.csv', sep='\t', header=True, index=False)
```

### Разбиение по дням(календарным)

```python
gp_table['Time_series'] = pd.to_datetime(gp_table['Date'], format='%d.%m.%Y', errors='ignore')
gp_table = gp_table.set_index('Time_series')
```

```python
bin_size = '10D'
gp_min = gp_table.resample(bin_size).min().dropna()
gp_max = gp_table.resample(bin_size).max().dropna()
gp_mean = gp_table.resample(bin_size).mean().dropna()
gp_std = gp_table.resample(bin_size).std().dropna()
```

```python
res_table = pd.concat([gp_min['MJD'],gp_mean['MJD'], gp_max['MJD'],gp_mean['W50, point'], gp_std['W50, point']], axis=1).dropna()
```

```python
res_table.columns = ['MJD_min', 'MJD_mean','MJD_max','W50_mean', 'W50_std'] 
```

```python
res_table
```

```python
res_table.to_csv('W50_bin_10_sigma.csv', sep='\t', header=True, index=False)
```

### Разбиение по календарным дням с использованием временного индекса для определения MJD 

```python
# Создание новой колонки
gp_table['DateTime'] = None
for idx, row in gp_table.iterrows():
    gp_table.loc[idx,'DateTime'] = str(row['Date']) + ' ' + str(row['Time start'])
```

```python
gp_table.head()
```

```python
gp_table['Time_series'] = pd.to_datetime(gp_table['DateTime'], format='%d.%m.%Y %H:%M:%S.%f', errors='ignore')
gp_table['Time_series'] -= pd.Timedelta('4 hour') # to UTC
gp_table = gp_table.set_index('Time_series')
```

```python
bin_size = '10D'
gp_mean = gp_table.resample(bin_size).mean().dropna()
gp_std = gp_table.resample(bin_size).std().dropna()
```

```python
res_table = pd.concat([gp_mean['W50, point'], gp_std['W50, point']], axis=1).dropna()
```

```python
res_table = res_table.reset_index()
```

```python
res_table['MJD_start'] = None
```

```python
for index, row in res_table.iterrows():
    time_utc_start = Time(str(row['Time_series']).replace(' ', 'T'), precision=7)
    res_table.loc[index,'MJD_start'] = str(time_utc_start.mjd)
```

```python
res_table.columns = ['Time_start' ,'W50_mean', 'W50_std', 'MJD_start']
res_table = res_table[['MJD_start', 'W50_mean', 'W50_std']]
```

```python
res_table
```

```python
res_table.to_csv('W50_bin_10.csv', sep='\t', header=True, index=False)
```

### Реализация алгоритма Максима
Сделать можно, например, вот так. У нас интервал 55400-58300, 2900 дней. Можно просто определять, попадает ли импульс в какой-то бин (MJD-55400)/10(30). Нацело делить на размер бина в днях, это целое число даст номер бина. Ну и для каждого бина заводишь массив, в который скидываешь w50, которые относятся к соотв. бину. После этого считаешь стат. величины для каждого такого массива, т.е. каждого бина.

```python
dect_res = {}
bin_size = 10
```

```python
for idx, row in gp_table.iterrows():
    bin_num = (row['MJD'] - 55400)//bin_size
    try:
        dect_res[bin_num].append(row['W50, point'])
    except KeyError:
        dect_res[bin_num] = []
        dect_res[bin_num].append(row['W50, point'])
```

```python
resul_tab = pd.DataFrame(columns=[
    '№ bin',
    'MJD_mean',
    'W50_mean',
    'STD',
    'len bin'])
```

```python
idx = 0
for key in dect_res.keys():
    resul_tab.loc[idx] = [
        key,    
        (key + 0.5)*bin_size + 55400,
        np.mean(dect_res[key]),
        np.std(dect_res[key]),
        np.size(dect_res[key])
    ]
    idx +=1
```

```python
resul_tab = resul_tab.sort_values(by=['№ bin'])
```

```python
resul_tab.head()
```

```python
resul_tab.to_csv('W50_bin_'
                 + str(bin_size)
                 + '_maxm_sigma.csv', sep='\t', header=True, index=False)
```
